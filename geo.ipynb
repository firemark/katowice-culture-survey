```python
import pandas as pd
import requests
```

```python
fm = pd.read_csv('survey-data.csv')
cities = set(fm.address_city)
len(cities)
```

```python
for city in cities:
    print city
```

```python
map_cities = {}
for city in cities:
    if not isinstance(city, str):
        continue
    city, _, _ = city.partition(',')
    print city, '...',
    resp = requests.get('https://nominatim.openstreetmap.org/search?city=Katowice&format=json', params={
        'city': city,
        'format': 'json',
    })
    
    try:
        resp.raise_for_status()
    except Exception as e:
        print 'omg!', e
        continue
    
    jj = resp.json()
    if len(jj) == 0:
        print 'not found! :('
        continue
    j = jj[0]
    map_cities[city] = {'lat': j['lat'], 'lon': j['lon']}
    print 'OK!'
```

```python
len(map_cities)
```

```python
import json
with open('geo.json', 'w') as fp:
    json.dump(map_cities, fp)
```
